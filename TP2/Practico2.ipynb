* Gender: genero del individuo.

* Weight: peso del individuo.
* CALC: ¿Con que frecuencia tomas alcohol?
* FAVC: ¿Ingeris comidas con alto indice calorico frecuentemente?

* TUE: ¿Cuanto tiempo usas dipositivos electronicos?
* CAEC: ¿Te alimentas con algo extra entre comidas?
* MTRANS: ¿Que transporte utilizas normalmente?
* NObeyesdad: Nivel de obesidad.
